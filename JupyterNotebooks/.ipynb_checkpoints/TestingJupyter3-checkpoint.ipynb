```python
import jupyros
import rospy
import ipywidgets as widgets
import math
from std_srvs.srv import SetBool, SetBoolRequest, SetBoolResponse
from rt2_assignment1.srv import Command, CommandRequest, CommandResponse
from geometry_msgs.msg import Twist
from nav_msgs.msg import Odometry

rospy.init_node( "jupyter_end_ui" )
name_ui_trigger = "/ui_trigger"
rospy.wait_for_service( name_ui_trigger )
cli_ui_trigger = rospy.ServiceProxy( name_ui_trigger, SetBool )
cli_ui_trigger( True )

tw = Twist( )
tw.linear.x = 0.0
tw.linear.y = 0.0
tw.linear.z = 0.0
tw.angular.x = 0.0
tw.angular.y = 0.0
tw.angular.z = 0.0
tw0 = Twist( )
tw0.linear.x = 0.0
tw0.linear.y = 0.0
tw0.linear.z = 0.0
tw0.angular.x = 0.0
tw0.angular.y = 0.0
tw0.angular.z = 0.0

pub_cmd_vel = rospy.Publisher( "/cmd_vel", Twist, queue_size=1000 )
pub_cmd_vel.publish( tw0 )

lay_item = widgets.Layout( 
    display="auto" )

lay_vitem = widgets.Layout( 
    display="auto", 
    align_self="center" )

lay_stop_button = widgets.Layout( 
    display="auto", 
    align_self="center", 
    width="75%", 
    margin="0" )

lay_auto_vcenter = widgets.Layout( 
    display="auto",
    position="absolute",
    top="50%" )

lay_vertical = widgets.Layout(
    display="flex", 
    flex_flow="column",
    align_items="stretch",
    justify_content="center" )

lay_horizontal = widgets.Layout(
    display="flex",
    flex_direction="column",
    justify_content="center",
    align_items="baseline" )

b_up = widgets.Button(description="^", layout=lay_item)
b_down = widgets.Button(description="v", layout=lay_item)
b_left = widgets.Button(description="<", layout=lay_vitem)
b_right = widgets.Button(description=">", layout=lay_vitem)
b_stop = widgets.Button(description="STOP", layout=lay_stop_button)
txt_front = widgets.Label( value="front: 0.00m/s", layout=lay_vitem )
txt_left = widgets.Label( value="left: 0.00m/s", layout=lay_vitem )
txt_rot = widgets.Label( value="rotation: 0.00rad/s", layout=lay_vitem )
txt_box = widgets.Box( [txt_front, txt_left, txt_rot], layout=lay_horizontal )
vertical_buttons = widgets.Box( [b_up, b_down], layout=lay_vertical )
cmd_pad = widgets.Box( [b_left, vertical_buttons, b_right], layout=lay_horizontal )
complete_pad = widgets.Box( [txt_box, cmd_pad, b_stop], layout=lay_vertical )

last_odom = Odometry( )
def n( num ):
    if math.abs(num) < 0.01:
        return 0
    else:
        return num

def get_odom( msg ):
    global last_odom
    last_odom = msg
    txt_front.value = ("front: {:.2f}m/s").format( n(msg.twist.twist.linear.x) )
    txt_left.value = ("left: {:.2f}m/s").format( n(msg.twist.twist.linear.y)  )
    txt_rot.value = ("rotation: {:.2f}rad/s").format( n(msg.twist.twist.angular.z) )
odom = rospy.Subscriber( "/odom", Odometry, get_odom )

linear_step = 0.5
angular_step = 0.25

def update( ):
    global txt_front, txt_left, txt_rot
    '''
    txt_front.value = ("front: {}m/s").format( tw.linear.x )
    txt_left.value = ("left: {}m/s").format( tw.linear.y  )
    txt_rot.value = ("rotation: {}rad/s").format( tw.angular.z )
    '''
    pub_cmd_vel.publish( tw )

def stop_all( b ):
    global tw, tw0
    tw.linear.x = 0.0
    tw.linear.y = 0.0
    tw.angular.z = 0.0
    update( )

def up( b ):
    global tw, linear_step
    tw.linear.x = tw.linear.x + linear_step
    update( )

def down( b ):
    global tw, linear_step
    tw.linear.x = tw.linear.x - linear_step
    update( )

def left( b ):
    global tw, angular_step
    tw.angular.z = tw.angular.z - angular_step
    update( )

def right( b ):
    global tw, angular_step
    tw.angular.z = tw.angular.z + angular_step
    update( )


b_stop.on_click( stop_all )
b_up.on_click( up )
b_down.on_click( down )
b_left.on_click( left )
b_right.on_click( right )
display( complete_pad )
```
